# Scraping Amazon Listings

Import Libraries- requests and BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36',
}
```

Here we are using the women's clothing page on Amazon. 

```python
response = requests.get("https://www.amazon.in/s?i=apparel&bbn=1968542031&rh=n%3A1968542031%2Cp_36%3A-39900%2Cp_72%3A1318476031%2Cp_85%3A10440599031&pf_rd_i=283155&pf_rd_i=6648217031&pf_rd_m=A1K21FY43GMZF8&pf_rd_m=A1K21FY43GMZF8&pf_rd_p=05b91ec0-2ed2-4cf5-a33a-23a7d8762cf1&pf_rd_p=preview%3Ahfdwidgetbuilderservice&pf_rd_r=35b4cd2c-f088-4d84-9397-bfbd0c4ca14a&pf_rd_r=JVQ41BHKZV7PAX3E612N&pf_rd_s=center-1&pf_rd_s=merchandised-search-4&pf_rd_t=101&pf_rd_t=101&qid=1647406728&rnid=10440598031&ref=is_r_p_85_1",headers=headers)
content = response.content
soup = BeautifulSoup(content,"html.parser")
```

From each listing, we are extracting the name, price, rating and number of reviews of the product.

```python
name_class="a-size-base-plus a-color-base a-text-normal"

rating_class="a-popover-trigger a-declarative"

reviews_class="a-size-base s-light-weight-text"

price_class="a-price-whole"
```

```python
Name=[]
for n in soup.findAll("span",{"class": name_class}):
    Name.append(n.get_text())

Price=[]
for p in soup.findAll("span",{"class": price_class}):
    Price.append(p.get_text())
    
Rating=[]
for r in soup.findAll("a",{"class": rating_class}):
    Rating.append(r.get_text())
    
Reviews=[]
for rev in soup.findAll("span",{"class": reviews_class}):
    Reviews.append(rev.get_text())
```

Storing the scraped data into a dataframe

```python
import pandas as pd
df=pd.DataFrame()
```

```python
df['Product Name']=Name
df['Price']=Price
df['Rating']=Rating
df['No of Reviews']= Reviews
```
